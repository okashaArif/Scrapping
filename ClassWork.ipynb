```python
import bs4
from bs4 import BeautifulSoup
import requests
```

```python
website= 'https://www.bbc.com/news/live/world-middle-east-67223217'
```

```python
result=requests.get(website)
```

```python
content=result.text
```

```python
soup=BeautifulSoup(content,'lxml')
```

```python
print(soup.prettify())
```

```python
headers=soup.findAll(['h1','h2','h3','h4'])
print (headers)
```

```python
namelist=soup.find('span',{'class':'lx-stream-post__header-text gs-u-align-middle'})

print(namelist.get_text())
```

```python
paragraphs = soup.find('div',{'class':'lx-stream-post-body'})
for paragraph in paragraphs.select('p'):
    print (paragraph.get_text())
```

```python
for i in soup.select('p'):
    print(i.text)
```
